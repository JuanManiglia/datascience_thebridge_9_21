### Step 4. Which continent drinks more beer on average?

```python
drinks.groupby(['continent'])['beer_servings'].mean().idxmax()
```

```python
df_grouped = drinks.groupby(['continent'], as_index=False).mean()
df_grouped
```

```python
df_grouped[df_grouped['beer_servings'] == df_grouped['beer_servings'].max()][['continent','beer_servings']]
```

```python
drinks['continent_fixed'] = np.where(drinks['continent'].isna(),'NorteAmerica', drinks['continent'])
# drinks['continent_fixed'] = np.where(drinks['continent_fixed']=='NorteAmerica', 'NA' , drinks['continent'])
drinks.head()
```

### Step 5. For each continent print the statistics for wine consumption.

```python
drinks.groupby(['continent_fixed'], as_index=False)[['wine_servings']].mean()
```

### Step 6. Print the mean alcohol consumption per continent for every column

```python
drinks.groupby(['continent_fixed']).mean()
```

### Step 7. Print the median alcohol consumption per continent for every column

```python
drinks.groupby('continent').median()
```
