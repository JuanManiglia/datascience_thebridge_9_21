### SOLUCION

```python
# Podria ser
chipo[chipo.item_price == max(chipo.item_price)][['item_name', 'item_price', 'quantity']].drop_duplicates()
```

```python
chipo["unity_price"] = chipo.item_price / chipo.quantity
chipo
```

```python
chipo_2 = chipo[chipo.unity_price == max(chipo.unity_price)]
chipo_2[['item_name', 'choice_description', 'unity_price']].drop_duplicates()
```
